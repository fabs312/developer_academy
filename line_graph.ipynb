```python
#import lib and packages
import numpy as np
import matplotlib.pyplot as plt

year, age_m, age_f = np.loadtxt('eg7-marriage-ages.txt', unpack=True, skiprows=3) #downloaded data in the same path where this code is

fig, ax = plt.subplots()

#Plot ages with male or female synbols as markers

ax.plot(year, age_m, marker='$u2642$', markersize=14, c='blue', lw=2, mfc='blue', mec='blue') #lw line width, , mec marker edge colour

ax.plot(year, age_f, marker='$u2640$', markersize=14, c='magenta', lw=2, mfc='magenta', mec='magenta')
ax.grid()

ax.set_xlabel('Year')
ax.set_ylabel('Age')
ax.set_title('Median age at first marriage in the US, 1890-2010')





```
