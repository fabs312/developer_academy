```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111) #111= 1x1 grid, first subplot????

cities = ['Boston', 'Houston', 'Detroit', 'San Jose', 'Phoenix']
# line styles: solid, dashes, dots, dash-dots, and dot-dot-dash
linestyles = [{'ls': '-'}, {'ls': '--'}, {'ls': ':'}, {'ls': '-.'},
              {'dashes': [2, 4, 2, 4, 8, 4]}]

for i, city in enumerate(cities): #for each ITEM in LIST: do something
    filename = '{}.tsv'.format(city.lower()).replace(' ', '_')
    yr, pop = np.loadtxt(filename, unpack=True) #load data from a text file, unpack=true retrieve column arrays individually
    line, = ax.plot(yr, pop/1.e6, label=city, color='k', **linestyles[i])

#building my graph
ax.legend(loc='upper left') #location of legend
ax.set_xlim(1800, 2020)
ax.set_xlabel('Year')
ax.set_ylabel('Population (millions)')
ax.set_title('Population ttrends for 5 US cities')
plt.show()
```
