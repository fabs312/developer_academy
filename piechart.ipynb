```python
#import lib and packages
import numpy as np
import matplotlib.pyplot as plt

#annual greenhouse gas emssions in 2007
gas_emissions = np.array([(r'$\mathrm{CO_2}$-d', 2.2),
                           (r'$\mathrm{CO_2}$-f', 8.0),
                           ('Nitorus\nOxide', 1.0),
                           ('Metahne', 2.3),
                           ('Halocarbons', 0.1)],
                           dtype=[('source', 'U17'), ('emission', 'f4')]) #the data type in the piechart methane is a unicode17 for strings and f4 for float 2.3

#5 coulours using hex colour
colors= ['#C7B299', '#A67C52', '#C69C6E', '#754C24', '#534741']

explode = [0, 0, 0.1, 0, 0]  #specifies the fraction i want of the pie chart

fig, ax = plt.subplots() #make axes 
ax.axis('equal')    # make a circle

ax.pie(gas_emissions['emission'], colors=colors, shadow=True, startangle=90, explode=explode, labels=gas_emissions['source'], autopct='%.1f%%', pctdistance=1.15, labeldistance=1.3)


```
